```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
data={"age group":['<21','21-24','24-27','27-30','30-33','33-36','36-39','39-42'],"Total number of loans":[9,310,511,4000,4568,5698,8209,8117],"NUmber of bad loans":[2,14,20,172,169,188,197,211],"Number of good loans":[7,296,491,3828,4399,5510,8012,7906],"Percentage of bad loans":['22.2%','4.5%','3.9%','4.3%','3.7%','3.3%','2.4%','2.6%'],"Percentage good loans":['77.8%','95.5%','96.1%','95.7%','96.3%','96.7%','97.6%','97.4%']}
df=pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['age group'], df['Total number of loans'], color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Total number of loans')
plt.title('Total number of loans by age group')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.bar(df['age group'], df['Number of good loans'], color='green', label='Good Loans')
plt.bar(df['age group'], df['NUmber of bad loans'], color='red', bottom=df['Number of good loans'], label='Bad Loans')
plt.xlabel('Age Group')
plt.ylabel('Number of Loans')
plt.title('Distribution of Good and Bad Loans by Age Group')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['Number of good loans'], df['NUmber of bad loans'],  c='red',  alpha=0.8)
plt.xlabel('Number of Good Loans')
plt.ylabel('Number of Bad Loans')
plt.title('Scatter Plot: Good Loans vs Bad Loans')
plt.grid(True)
plt.colorbar(label='Number of Loans')
plt.show()

```
