```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
    "value":[10,15,22,28,35]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel('Date')
plt.ylabel("Value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```

```python
data={
    "Date": pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01","2022-11-01","2022-12-01"]),
    "Value":[10,30,15,42,39,45,25,28,60,66,44,36]
}
df= pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"], marker='o',color="green")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/sunspotarea.csv', parse_dates=['date'], index_col='date')

# Seasonal decomposition
result = seasonal_decompose(data, model='additive', period=12)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 8), dpi=100)

# Original data
data.plot(ax=axes[0, 0], title='Original Data', legend=False)

# Trend component
result.trend.plot(ax=axes[0, 1], title='Trend Only', legend=False)

# Seasonal component
result.seasonal.plot(ax=axes[1, 1], title='Seasonal Only', legend=False)

# Time component
result.resid.plot(ax=axes[1, 0], title='Residuals', legend=False)

plt.tight_layout()
plt.show()
```
